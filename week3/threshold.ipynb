```python
import pandas as pd
```

```python
synonyms_output = pd.read_csv('synonyms_output.txt', sep = '|', header = None, names = ['line'])
```

```python
synonyms_output.head()
```

```python
synonyms_list = []
for index, row in synonyms_output.iterrows():
#     print(row['line'])
    line = row['line']
    if 'Query word?' in line:
#         print(line)
        line = line[11:]
#         print(line)
    synonyms_list.append(line)
```

```python
synonyms_list
```

```python
with open("synonyms_output", 'w') as file:
    file.write('\n'.join(synonyms_list))
```

```python
lowest = 1.0
print(type(lowest))
import re
for word in synonyms_list:
    numbers = re.findall("\d+\.\d+", word)
    if len(numbers) == 1:
        number = float(numbers[0])
        if  number < lowest:
            lowest = number
```
